```python
import pandas as pd
import re
```

```python
df = pd.read_json('../ZDF/zdfscraping/zdfscraping/spiders/zdf_dataset_version_1')
print(len(df))
df = df[df["Text"].apply(lambda x: len(x) > 0)]
print(len(df))
df.head()
```

```python
df['Link'] = df['Link'].str.replace(r'<HtmlResponse 200 (https?://[^\s>]+)>', r'\1', regex=True)

def extract_date_time(text):
    pattern = r'(\d{2}\.\d{2}\.\d{4}) \|\n\s*(\d{2}:\d{2})'
    match = re.search(pattern, text)
    if match:
        return match.group(1), match.group(2)
    return None, None

df[['date', 'time']] = df['publish_time'].apply(lambda x: pd.Series(extract_date_time(x[0])))

df.head()
```
